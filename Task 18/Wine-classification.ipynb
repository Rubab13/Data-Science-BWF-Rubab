<div>
    <h2 style='color:#0B4F6C;font-weight:bold;font-size:1.5rem;padding:1rem 0;'>Importing Libraries</h2>
</div>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

<div>
    <h2 style='color:#0B4F6C;font-weight:bold;font-size:1.5rem;padding:1rem 0;'>Importing Data</h2>
</div>

Link to Dataset: https://www.kaggle.com/datasets/nareshbhat/wine-quality-binary-classification

```python
data = pd.read_csv("wine.csv")
```

```python
data.head()
```

***

<div>
    <h2 style='color:#0B4F6C;font-weight:bold;font-size:1.5rem;padding:1rem 0;'>Separating Features and Label</h2>
</div>

```python
x = data.iloc[:, 0:11]
```

```python
y = data.iloc[:, 11]
```

<div>
    <h2 style='color:#0B4F6C;font-weight:bold;font-size:1.5rem;padding:1rem 0;'>Features</h2>
</div>

```python
x
```

<div>
    <h2 style='color:#0B4F6C;font-weight:bold;font-size:1.5rem;padding:1rem 0;'>Label</h2>
</div>

```python
y
```

***

<div>
    <h2 style='color:#0B4F6C;font-weight:bold;font-size:1.5rem;padding:1rem 0;'>Separating Training and Testing Data</h2>
</div>

```python
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size = 0.2)
```

<div>
    <h2 style='color:#0B4F6C;font-weight:bold;font-size:1.5rem;padding:1rem 0;'>Creating the model</h2>
</div>

```python
model = SVC(kernel='linear')
```

<div>
    <h2 style='color:#0B4F6C;font-weight:bold;font-size:1.5rem;padding:1rem 0;'>Training the model on data</h2>
</div>

```python
model.fit(xtrain, ytrain)
```

<div>
    <h2 style='color:#0B4F6C;font-weight:bold;font-size:1.5rem;padding:1rem 0;'>Making Predictions with the model</h2>
</div>

```python
yprediction = model.predict(xtest)
```

<div>
    <h2 style='color:#0B4F6C;font-weight:bold;font-size:1.5rem;padding:1rem 0;'>Accuracy Score of the model</h2>
</div>

```python
print("Accuracy:", accuracy_score(ytest, yprediction))
```

***

<div>
    <h2 style='color:#0B4F6C;font-weight:bold;font-size:1.5rem;padding:1rem 0;'>Testing the model on new data</h2>
</div>

```python
new_data_point = np.array([[7.4, 0.70, 0.00, 1.9, 0.076, 11.0, 34.0, 0.9978, 3.51, 0.56, 9.4]])
```

```python
prediction = model.predict(new_data_point)
```
