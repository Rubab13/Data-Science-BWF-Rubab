```python
import pandas as pd
import numpy as np
```

<div>
    <h2 style='color:#0B4F6C;font-weight:bold;font-size:1.5rem;padding:1rem 0;'>Importing Data</h2>
</div>

The dataset I have selected is already clean. So there is no pre processing involved. I will directly move on to Model Development that involve the machine learning concepts.

```python
data = pd.read_csv("Iris.csv")
```

```python
data
```

***

```python
df = data.values
```

<div>
    <h2 style='color:#0B4F6C;font-weight:bold;font-size:1.5rem;padding:1rem 0;'>Separating Features and Label</h2>
</div>

I will select all rows but only the 4 necessary features. The sepal length, sepal width, petal length and the petal width.

```python
x = df[:, 1:5]
```

```python
x
```

```python
y = df[:, 5]
```

```python
y
```

***

<div>
    <h2 style='color:#0B4F6C;font-weight:bold;font-size:1.5rem;padding:1rem 0;'>Separating Training and Testing Data</h2>
</div>

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size = 0.2)
```

***

<div>
    <h2 style='color:#0B4F6C;font-weight:bold;font-size:1.5rem;padding:1rem 0;'>Model Creation</h2>
</div>

```python
from sklearn.svm import SVC
```

```python
model = SVC(kernel='linear')
```

<div>
    <h2 style='color:#0B4F6C;font-weight:bold;font-size:1.5rem;padding:1rem 0;'>Model Fitting over Data</h2>
</div>

```python
model.fit(xtrain, ytrain)
```

<div>
    <h2 style='color:#0B4F6C;font-weight:bold;font-size:1.5rem;padding:1rem 0;'>Model Prediction</h2>
</div>

```python
ypred = model.predict(xtest)
```

<div>
    <h2 style='color:#0B4F6C;font-weight:bold;font-size:1.5rem;padding:1rem 0;'>Accuracy Measurement</h2>
</div>

```python
from sklearn.metrics import accuracy_score
```

```python
print("Accuracy:", accuracy_score(ytest, ypred)*100)
```

***

<div>
    <h2 style='color:#0B4F6C;font-weight:bold;font-size:1.5rem;padding:1rem 0;'>Testing over new Data</h2>
</div>

I took the first row of my dataset to find out the input and yes the output came out correct. The first row belonged to the Setosa group.

```python
new_data_point = np.array([[5.1, 3.5, 1.4, 0.2]])  # Example: Sepal length, Sepal width, Petal length, Petal width

prediction = model.predict(new_data_point)

print("Predicted class:", prediction[0])
```
